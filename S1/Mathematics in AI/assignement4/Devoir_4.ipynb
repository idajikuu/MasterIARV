```python
import random
from math import *
def monte_carlo_pi(n):
    dots = 0
    for i in range(n):
        x, y = random.random(), random.random() 
        if x*x + y*y <= 1: 
            dots += 1
    return 4 * dots / n
value=monte_carlo_pi(10000)
print("Value of pi by the Monte carlo simulation :",value)
print("Error ",pi-value)
```

```python
#by samples
from math import *
import random
def function(x):
  return 1/np.sqrt(1+x*x)
N=0
counter=0
while(N<10000):
  x=random.uniform(-3,7)
  y=random.uniform(0,1)
  if y<function(x):
    counter=counter+1
  N=N+1
b=7
a=-3

rec_area=(b-a)*1
print(counter/N * rec_area)
```

```python
def fun(x):
  return 1/np.sqrt(1+x*x)
def rect(function,lowerBound, upperBound, size):
  sample_list = []
  while len(sample_list) < size: 
    sample_list.append(function(random.uniform(lowerBound,upperBound)))
  return [sum(sample_list) * ((upperBound - lowerBound)/size),sample_list]
```

```python
integral_estimate, list_sample = rect(fun, -3,7, 10000)
```
